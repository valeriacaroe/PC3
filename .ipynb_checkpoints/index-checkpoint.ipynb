```python
import os, pandas as pd
reserve=pd.read_csv("https://github.com/valeriacaroe/PC3/raw/refs/heads/main/Reserves%20of%20foreign%20exchange%20and%20gold.csv")
energy=pd.read_csv("https://github.com/valeriacaroe/PC3/raw/refs/heads/main/Energy%20consumption%20per%20capita.csv")
electricity=pd.read_csv("https://github.com/valeriacaroe/PC3/raw/refs/heads/main/Electricity%20-%20installed%20generating%20capacity.csv")
education=pd.read_csv("https://github.com/valeriacaroe/PC3/raw/refs/heads/main/Education%20expenditures.csv")

reserve.head()
```

```python
# Elimino columnas que repiten información o tienen información que no usare en esta oportunidad
reserve1=reserve.drop(['slug','ranking'], axis=1)
energy1=energy.drop(['slug','ranking'], axis=1)
electricity1=electricity.drop(['slug','ranking'], axis=1)
education1=education.drop(['slug','ranking'], axis=1)
```

🧹 Limpieza y formateo de datos de data frames individualmente de acuerdo al caso

```python
reserve1['value'] = reserve1['value'].astype(str).str.replace(',', '', regex=False)
reserve1['value'] = reserve1['value'].str.replace('[^0-9,]', '', regex=True)
reserve1['value'] = pd.to_numeric(reserve1['value'], errors='coerce')
reserve1.info()
```

```python
energy1[' Btu/person'] = energy1[' Btu/person'].astype(str).str.replace(',', '', regex=False)
energy1[' Btu/person'] = pd.to_numeric(energy1[' Btu/person'], errors='coerce')
energy1.columns = energy1.columns.str.replace(' ', '', regex=False)
energy1.info()
```

```python
electricity1[' kW'] = electricity1[' kW'].astype(str).str.replace(',', '', regex=False)
electricity1[' kW'] = pd.to_numeric(electricity1[' kW'], errors='coerce')
electricity1.columns = electricity1.columns.str.replace(' ', '', regex=False)
electricity1.info()
```

```python
education1.info()
```

🧠 Merge inicial entre dos data frames y análisis de datos pérdidos antes de merge final

```python
merge1=education1.merge(reserve1,left_on='name',right_on='name')
merge1
```

```python
merge2=electricity1.merge(energy1,left_on='name',right_on='name')
merge2
```

```python
only_merge1=set(merge1.name)-set(merge2.name)
only_merge1
```

```python
only_merge2=set(merge1.name)-set(merge2.name)
only_merge2
```

```python
mergef=merge1.merge(merge2,left_on='name',right_on='name')
mergef
```

📊 Reporte del promedio, valor mínimo y máximo para cada variable por región

```python
education_report = mergef.groupby('region_x_x')['% of GDP'].agg(['mean', 'max', 'min']).reset_index()
print(education_report)
```

```python
reserve_report = mergef.groupby('region_y_x')['value'].agg(['mean', 'max', 'min']).reset_index()
print(reserve_report)
```

```python
electricity_report = mergef.groupby('region_x_y')['kW'].agg(['mean', 'max', 'min']).reset_index()
print(electricity_report)
```

```python
energy_report = mergef.groupby('region_y_y')['Btu/person'].agg(['mean', 'max', 'min']).reset_index()
print(energy_report)
```

✏️ Obtención de valores previos para intervalos y creación de categorías por variable

```python
#Búsqueda de valores mínimos y máximos globalmente para cada variable

reserve_min_global = reserve_report['min'].min()
reserve_max_global = reserve_report['max'].max()

print(f"RESERVE")
print(f"El valor mínimo global es: {reserve_min_global}")
print(f"El valor máximo global es: {reserve_max_global}")

energy_min_global = energy_report['min'].min()
energy_max_global = energy_report['max'].max()

print(f"\nENERGY")
print(f"El valor mínimo global es: {energy_min_global}")
print(f"El valor máximo global es: {energy_max_global}")

electricity_min_global = electricity_report['min'].min()
electricity_max_global = electricity_report['max'].max()

print(f"\nELECTRICITY")
print(f"El valor mínimo global es: {electricity_min_global}")
print(f"El valor máximo global es: {electricity_max_global}")

education_min_global = education_report['min'].min()
education_max_global = education_report['max'].max()

print(f"\nEDUCATION")
print(f"El valor mínimo global es: {education_min_global}")
print(f"El valor máximo global es: {education_max_global}")
```

```python
# Valores de intervalo para cada variable de acuerdo al valor mínimo y máximo
int_reserve=(3265000000000-30450000)/5
print(int_reserve)

int_energy=(814308000-649000)/5
print(int_energy)

int_electricity=(2949000000-29000)/5
print(int_electricity)

int_education=(12.8-0.3)/5
print(int_education)
```

```python
mergef['education_score']=pd.cut(mergef['% of GDP'],
                           bins=[0.3,2.8,5.3,7.8,10.3,12.8],
                           labels=['very bad','bad','average','good','very good'],
                           ordered=True)

mergef['electricity_score']=pd.cut(mergef['kW'],
                           bins=[29000, 589823200, 1179617400, 1769411600, 2359205800, 2949000000],
                           labels=['very bad','bad','average','good','very good'],
                           ordered=True)

mergef['energy_score']=pd.cut(mergef['Btu/person'],
                           bins=[649000, 163380800, 326112600, 488844400, 651576200, 814308000],
                           labels=['very bad','bad','average','good','very good'],
                           ordered=True)

mergef['reserve_score']=pd.cut(mergef['value'],
                           bins=[30450000, 653024360000, 1306018270000, 1959012180000, 2612006090000, 3265000000000],
                           labels=['very bad','bad','average','good','very good'],
                           ordered=True)
mergef.head()
```

⏳ Arreglos finales para data frame final solicitado

```python
mergef = mergef.drop(['region_y_x', 'region_x_y', 'region_y_y'], axis=1)
mergef = mergef.drop(['date_of_information_x_x', 'date_of_information_x_y', 'date_of_information_y_x', 'date_of_information_y_y'], axis=1)
mergef = mergef.rename(columns={'region_x_x': 'region'})
mergef = mergef.rename(columns={'name': 'country'})
mergef = mergef.rename(columns={'value': 'value ($)'})
mergef = mergef.rename(columns={'kW': 'electricity (kW)'})

mergef.head()
```

```python
columna=mergef.pop('region')
mergef.insert(1, 'region', columna)
mergef
```
